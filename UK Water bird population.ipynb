```python
import pandas as pd 
import plotly.express as px
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
df = pd.read_csv(r'C:\Users\Thoma\Downloads\sds-waterbirds-201003-1.csv')
```

```python
keys = [i for i in range(63)]

all_dataframes = {}
for i, county in zip(keys, df['Winter'].unique()):
    df2 = df[df['Winter'] == county].sort_values('Winter', ascending=True)
    all_dataframes[i] = df2 
    
master_df = []
for i in all_dataframes:
    master_df.append(all_dataframes[i])
    
df_all = pd.concat(master_df)
```

```python
df_all
```

```python
plot_df = df_all
```

```python
birdspecies = ['All species (46 species)', 'Wildfowl (27 species)', 'Wader (15 species)']
```

```python
fig = px.bar(plot_df, x='Location', y=birdspecies, animation_frame='Winter', 
             labels={'variable':'type of birds', 'value':'number of birds'}, 
             barmode='group',range_y=[0, 230], title='UK water bird population')
fig.show()
```
