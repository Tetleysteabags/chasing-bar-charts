```python
keys = [i for i in range(63)]

all_dataframes = {}
for i, county in zip(keys, df['Provider County'].unique()):
    df2 = df[df['Provider County'] == county].sort_values('Year', ascending=True)
    all_dataframes[i] = df2 
    
master_df = []
for i in all_dataframes:
    master_df.append(all_dataframes[i])
    
df_all = pd.concat(master_df)
```

```python
df_all
```

```python
selected_counties = df_all['Provider County'].sample(10, replace=True).values.tolist()
selected_counties
```

```python
plot_df = df_all[df_all['Provider County'].isin(selected_counties)]
plot_df
```

```python
plot_df['Observed Prescribing Rate per 100 Visits'].max()
```

```python
plot_df['Observed Prescribing Rate per 100 Visits'].min()
```

```python
fig = px.bar(plot_df, x='Provider County', y="Observed Prescribing Rate per 100 Visits", color="Provider County",
  animation_frame="Year", animation_group="Provider County", range_y=[25, 80], 
             title='Antibiotic Prescribing rates in US counties')
fig.show()
```
